```python
import geopandas as gpd
```

```python
def load_tram_metro_data():
    url = "https://maps.amsterdam.nl/open_geodata/geojson_latlng.php?KAARTLAAG=TRAMMETRO_PUNTEN_2025&THEMA=trammetro"
    gdf = gpd.read_file(url)
    return gdf

tram_metro_gdf = load_tram_metro_data()
tram_metro_gdf
```

```python
for _, row in tram_metro_gdf.iterrows():
        # print(row)
        # print(row.geometry)
        print(row.geometry.coords[0])  
```
